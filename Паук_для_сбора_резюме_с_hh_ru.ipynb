```python
import scrapy
import time
import csv
```

```python
class ResumeSpider(scrapy.Spider):
    name = "ResumeSpider"
    start_urls = ["https://hh.ru/search/resume"]

    def parse(self, response):
        links = response.css("div.resume-search-item__header a::attr(href)")
        for link in links:
            time.sleep(3)
            yield response.follow(link, self.parse_сheese)

        link = response.css("div.pager a::attr(href)")[-1].get()
        yield response.follow(link, self.parse)

    def parse_сheese(self, response):
        yield {
            "name": response.css('span.resume-block__title-text::text').get(),
            "salary": response.css('span.resume-block__salary::text').get(),
            "experience": response.css('h2 span.resume-block__title-text_sub span::text').get(),
            "key_skills": response.css('div.bloko-tag-list span::text').getall()
        }


```
